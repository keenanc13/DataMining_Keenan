# Handling Missing Values and Unique Values

```python
# Cek null values pada variabel property_type
display(df['property_type'].value_counts())


print("\nNull count for 'property_type':")
display(df['property_type'].isnull().sum())
```

```python
#diisi dengan value 'rumah'
df['property_type'] = df['property_type'].fillna('rumah')

display(df['property_type'].value_counts())
display(df['property_type'].isnull().sum())
```

```python
# Fill missing values in 'land_size_m2' with the median
median_land_size = df['land_size_m2'].median()
df['land_size_m2'].fillna(median_land_size, inplace=True)

# Fill missing values in 'building_size_m2' with the median
median_building_size = df['building_size_m2'].median()
df['building_size_m2'].fillna(median_building_size, inplace=True)

# Verify that missing values have been filled
print("Missing values after filling:")
display(df[['land_size_m2', 'building_size_m2']].isnull().sum())
```

```python
# Fill missing values in 'ads_id' with 'unknown'
df['ads_id'] = df['ads_id'].fillna('unknown')

# Fill missing values in 'floors' with the mode
mode_floors = df['floors'].mode()[0] # .mode() returns a Series, take the first element
df['floors'] = df['floors'].fillna(mode_floors)

# Verify that missing values have been filled
print("Missing values after filling:")
display(df[['ads_id', 'floors']].isnull().sum())
```

```python
print("Examining 'bathrooms':")
display(df['bathrooms'].unique())
display(df['bathrooms'].value_counts(dropna=False).head()) # Include NaN in counts
display(df['bathrooms'].describe())
print("\nNull count for 'bathrooms':")
display(df['bathrooms'].isnull().sum())

print("\nExamining 'bedrooms':")
display(df['bedrooms'].unique())
display(df['bedrooms'].value_counts(dropna=False).head()) # Include NaN in counts
display(df['bedrooms'].describe())
print("\nNull count for 'bedrooms':")
display(df['bedrooms'].isnull().sum())
```

```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np

features = ["building_size_m2", "land_size_m2", "floors"]

# Fill missing values for 'bedrooms'
mask_bed_missing = df["bedrooms"].isna()
train_bed = df[df["bedrooms"].notna()]
test_bed = df[df["bedrooms"].isna()]

if not test_bed.empty:
    model_bed = DecisionTreeRegressor(random_state=42)
    model_bed.fit(train_bed[features], train_bed["bedrooms"])
    preds_bed = model_bed.predict(test_bed[features])
    df.loc[mask_bed_missing, "bedrooms"] = np.round(preds_bed).astype(int)

print("Filled 'bedrooms' missing values:")
display(df.loc[mask_bed_missing, ["bedrooms"] + features])


# Fill missing values for 'bathrooms'
mask_bath_missing = df["bathrooms"].isna()
train_bath = df[df["bathrooms"].notna()]
test_bath = df[df["bathrooms"].isna()]

if not test_bath.empty:
    model_bath = DecisionTreeRegressor(random_state=42)
    model_bath.fit(train_bath[features], train_bath["bathrooms"])
    preds_bath = model_bath.predict(test_bath[features])
    df.loc[mask_bath_missing, "bathrooms"] = np.round(preds_bath).astype(int)

print("\nFilled 'bathrooms' missing values:")
display(df.loc[mask_bath_missing, ["bathrooms"] + features])

# Verify that missing values have been filled
print("\nMissing values after filling:")
display(df[['bedrooms', 'bathrooms']].isnull().sum())
```

```python
# Fill missing values in certificate, property_condition, and furnishing with 'unknown'
for col in ['certificate', 'property_condition', 'furnishing']:
    df[col].fillna('unknown', inplace=True)

# Verify missing values are filled
print("Missing values after filling:")
display(df[['certificate', 'property_condition', 'furnishing']].isnull().sum())
```

```python
# Move 'semi furnished' and 'unfurnished' from property_condition to furnishing
# Ensure they are not already in furnishing to avoid duplicates if running multiple times
furnishing_values_to_move = ['semi furnished', 'unfurnished']

for index, row in df.iterrows():
    if row['property_condition'] in furnishing_values_to_move:
        # If the value in furnishing is already 'unknown', replace it
        if row['furnishing'] == 'unknown':
            df.loc[index, 'furnishing'] = row['property_condition']
        # If the value is not unknown and not one of the values to move, append it (optional, depending on desired behavior)
        # else:
        #     df.loc[index, 'furnishing'] = str(row['furnishing']) + ', ' + str(row['property_condition'])

        # Replace the value in property_condition with 'unknown' or another placeholder
        df.loc[index, 'property_condition'] = 'unknown' # Or a more appropriate placeholder


# Move 'baru' from furnishing to property_condition
# Ensure 'baru' is not already in property_condition
furnishing_value_to_move_to_condition = 'baru'

for index, row in df.iterrows():
    if row['furnishing'] == furnishing_value_to_move_to_condition:
         # If the value in property_condition is already 'unknown', replace it
        if row['property_condition'] == 'unknown':
             df.loc[index, 'property_condition'] = row['furnishing']
         # If the value is not unknown and not 'baru', append it (optional)
         # else:
         #    df.loc[index, 'property_condition'] = str(row['property_condition']) + ', ' + str(row['furnishing'])

        # Replace the value in furnishing with 'unknown' or another placeholder
        df.loc[index, 'furnishing'] = 'unknown' # Or a more appropriate placeholder


# Verify the changes in unique values
print("\nUnique values after mapping:")
print("Property Condition:", df['property_condition'].unique())
print("Furnishing:", df['furnishing'].unique())
```

```python
# Drop the 'year_built' column
df = df.drop('year_built', axis=1)

# Fill missing values in 'building_age' with the median
median_building_age = df['building_age'].median()
df['building_age'] = df['building_age'].fillna(median_building_age)

# Fill missing values in 'building_orientation' with 'unknown'
df['building_orientation'] = df['building_orientation'].fillna('unknown')

# Verify that missing values have been handled
print("Missing values after handling:")
display(df[['building_age', 'building_orientation']].isnull().sum())
```

```python
# Remove ' mah' suffix first
df['electricity'] = df['electricity'].astype(str).str.replace(' mah', '', regex=False)

# Replace 'lainnya' (which was 'lainnya mah') with NaN
df['electricity'] = df['electricity'].replace('lainnya', np.nan)

# Convert 'electricity' to numeric, coercing errors to NaN
df['electricity'] = pd.to_numeric(df['electricity'], errors='coerce')

# Calculate the median of the 'electricity' column (excluding NaN)
median_electricity = df['electricity'].median()

# Fill the NaN values with the median
df['electricity'] = df['electricity'].fillna(median_electricity)

# Convert the column to integer type
df['electricity'] = df['electricity'].astype(int)

# Verify the changes
print("Unique values in 'electricity' after handling:")
display(df['electricity'].unique())
print("\nMissing values in 'electricity' after handling:")
display(df['electricity'].isnull().sum())
```

*   Mengurangi pengaruh nilai ekstrem yang mungkin merupakan data tidak akurat atau anomali.

DataFrame `df_filtered` sekarang berisi data setelah outlier tersebut dipotong, menghasilkan dataset yang lebih bersih dan representatif untuk analisis lebih lanjut.

# Feature Engineering

```python
# Custom binning with descriptive labels
binning_configs = {
    'price_in_rp': {
        'bins': [0, 500000000, 1500000000, 3000000000, df_filtered['price_in_rp'].max()],
        'labels': ['Budget', 'Mid-range', 'Premium', 'Luxury'] # Adjusted labels based on bins
    },
    'building_age': {
        'bins': [df_filtered['building_age'].min() -1 , 0, 5, 15, 30, df_filtered['building_age'].max()],
        'labels': ['Baru', 'Relatif Baru', 'Lama', 'Sangat Lama', 'Usia Tua']
    },
    'land_size_m2': {
        'bins': [df_filtered['land_size_m2'].min()-1, 50, 100, 200, 500, df_filtered['land_size_m2'].max()],
        'labels': ['Kompak', 'Standar', 'Luas', 'Sangat Luas', 'Ekstra Luas']
    },
    'building_size_m2': {
        'bins': [df_filtered['building_size_m2'].min()-1, 50, 100, 200, 500, df_filtered['building_size_m2'].max()],
        'labels': ['Kecil', 'Sedang', 'Besar', 'Sangat Besar', 'Ekstra Besar']
    },
     'electricity': {
        'bins': [df_filtered['electricity'].min()-1, 1300, 3500, 7700, df_filtered['electricity'].max()],
        'labels': ['Rendah', 'Standar', 'Tinggi', 'Sangat Tinggi']
     }
}

for col, config in binning_configs.items():
    binned_col_name = f'{col}_category'
    df_filtered[binned_col_name] = pd.cut(df_filtered[col], bins=config['bins'], labels=config['labels'], include_lowest=True)

    # Display the value counts for the new binned column
    print(f"Value counts for '{binned_col_name}':")
    display(df_filtered[binned_col_name].value_counts().sort_index())
    print("-" * 30)

# Display the first few rows with the new binned columns
print("\nDataFrame with new binned columns:")
display(df_filtered[[col for col in df_filtered.columns if '_category' in col]].head())
```

# Penjelasan Binning

Tujuan dari binning ini adalah untuk:

*   **Menyederhanakan data:** Mengurangi jumlah nilai unik dalam kolom numerik yang kontinu.
*   **Mengurangi pengaruh outlier:** Nilai ekstrem akan masuk ke dalam bin yang sama dengan nilai-nilai lain dalam rentang tersebut, mengurangi dampak individualnya.
*   **Mengubah distribusi data:** Dapat membantu dalam menangani data yang sangat miring (skewed).
*   **Memenuhi persyaratan model:** Beberapa algoritma machine learning bekerja lebih baik dengan data kategorikal atau binned.

Setelah proses binning, kolom-kolom numerik tersebut sekarang direpresentasikan dalam bentuk kategori (misalnya, 'Budget', 'Mid-range', 'Premium' untuk harga) yang lebih mudah diinterpretasikan dan dapat digunakan untuk analisis lebih lanjut atau sebagai fitur dalam model yang membutuhkan data kategorikal.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the list of binned columns
binned_columns = [col for col in df_filtered.columns if '_category' in col]

# Create bar plots for each binned column
plt.figure(figsize=(len(binned_columns) * 4, 10)) # Adjust figure size based on number of plots
for i, col in enumerate(binned_columns):
    plt.subplot(1, len(binned_columns), i + 1)
    sns.countplot(data=df_filtered, x=col, hue=col, palette='viridis', legend=False) # Add hue and legend=False
    plt.title(f'Distribution of {col}')
    plt.xlabel(col) # X-label for vertical plot
    plt.ylabel('Count') # Y-label for vertical plot
    plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability

plt.tight_layout()
plt.show()
```
